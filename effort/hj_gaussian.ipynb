```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data_g=pd.read_excel('C:/Users/user/Desktop/psl/fault_feature_jirak.xls')
data_l=pd.read_excel('C:/Users/user/Desktop/psl/fault_feature_seongan.xls')
data_n=pd.read_excel('C:/Users/user/Desktop/psl/feature_normal.xls')

# G=data_g.drop(['target','type','m'],axis=1)
# L=data_l.drop(['target','type','m'],axis=1)
# N=data_n.drop(['target','type','m'],axis=1)

G=data_g.filter(['VA_am','IA_am','VB_am','IB_am','VA_bm','IA_bm','VB_bm','IB_bm','VA_cm','IA_cm','VB_cm','IB_cm','VC_am','IC_am','VD_am','ID_am','VC_bm','IC_bm','VD_bm','ID_bm','VC_cm','IC_cm','VD_cm','ID_cm'])
L=data_l.filter(['VA_am','IA_am','VB_am','IB_am','VA_bm','IA_bm','VB_bm','IB_bm','VA_cm','IA_cm','VB_cm','IB_cm','VC_am','IC_am','VD_am','ID_am','VC_bm','IC_bm','VD_bm','ID_bm','VC_cm','IC_cm','VD_cm','ID_cm'])
N=data_n.filter(['VA_am','IA_am','VB_am','IB_am','VA_bm','IA_bm','VB_bm','IB_bm','VA_cm','IA_cm','VB_cm','IB_cm','VC_am','IC_am','VD_am','ID_am','VC_bm','IC_bm','VD_bm','ID_bm','VC_cm','IC_cm','VD_cm','ID_cm'])

GLN = pd.concat([G, L, N])

print(G.shape)
print(L.shape)
print(N.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy.linalg as linalg

n_components = 8

pca=PCA(n_components=n_components)
G_pca=pca.fit(GLN)

G = G_pca.transform(G)
L = G_pca.transform(L)
N = G_pca.transform(N)

columns = []

for i in range(n_components):
    pca_comp = 'pca_comp'
    n = str(i+1)
    columns.append(pca_comp+n)

G = pd.DataFrame(G, columns = columns)
L = pd.DataFrame(L, columns = columns)
N = pd.DataFrame(N, columns = columns)

G_mean=G.mean(0)
L_mean=L.mean(0)
N_mean=N.mean(0)
G_std=np.array(np.std(G))
L_std=np.array(np.std(L))
N_std=np.array(np.std(N))
print(G_mean)
print(L_mean)
print(N_mean)
```

```python
G_cov=G.cov()
L_cov=L.cov()
N_cov=N.cov()
print(G_cov)
print(G_cov.shape)
# G_V, G_W =np.linalg.eig(np.cov(G.values.T))
# L_V, L_W =np.linalg.eig(np.cov(L.values.T))
# N_V, N_W =np.linalg.eig(np.cov(N.values.T))

# print("Eigen value: ",G_V)
# print("Eigen vector: ",G_W)

```

```python
# 다변수에 위에꺼 넣기
# 지락확률분포 하나 선간확률분포 하나 정상확률분포 하나
from scipy.stats import multivariate_normal
g_pdf=multivariate_normal(G_mean,G_cov,allow_singular=True)
n_pdf=multivariate_normal(N_mean,N_cov,allow_singular=True)
l_pdf=multivariate_normal(L_mean,L_cov,allow_singular=True)
```

```python
test=pd.read_excel('C:/Users/user/Desktop/psl/fault_feature_seongan.xls')
data_te=test.filter(['VA_am','IA_am','VB_am','IB_am','VA_bm','IA_bm','VB_bm','IB_bm','VA_cm','IA_cm','VB_cm','IB_cm','VC_am','IC_am','VD_am','ID_am','VC_bm','IC_bm','VD_bm','ID_bm','VC_cm','IC_cm','VD_cm','ID_cm'])
print(data_te.shape)

data_te = G_pca.transform(data_te)

data_te = pd.DataFrame(data_te, columns=columns)

prob_g = g_pdf.pdf(data_te) # g_pdf.pdf(48차원 데이터) -> 해당 데이터의 지락고장 확률이 나온다.
prob_n = n_pdf.pdf(data_te)
prob_l = l_pdf.pdf(data_te)
print(prob_g)
print(prob_n)
print(prob_l)
```
